```python
# prompt: make a heat map sub_category1 and sub_category2 with business_servie in those have highest urgency for tickets

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'urgency' column exists and represents urgency level
# Replace 'urgency' with your actual urgency column name if different

# Filter for business_service tickets with highest urgency
highest_urgency_tickets = df[df['business_service'] == True].sort_values('urgency', ascending=False)

# Create a pivot table for the heatmap
heatmap_data = highest_urgency_tickets.pivot_table(index='sub_category1', columns='sub_category2', values='urgency', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(12, 8))  # Adjust figure size as needed
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Urgency of Business Service Tickets by Sub-Category')
plt.show()
```

```python
# prompt: make heat map between business_type and priority with queue

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'df'
# and it contains columns 'business_type', 'priority', and 'queue'

# Create the pivot table
heatmap_data = df2.pivot_table(index='business_type', columns='priority', values='queue', aggfunc='count', fill_value=0)

# Create the heatmap
plt.figure(figsize=(13, 5))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Business Type vs. Priority with Queue Counts')
plt.xlabel('Priority')
plt.ylabel('Business Type')
plt.show()
```

```python
# prompt: which language has the most problem related to it services and which has most with tech online store and which least with all

import pandas as pd

# Load the dataset
df2 = pd.read_csv("/content/drive/MyDrive/Datasets/helpdesk_customer_multi_lang_tickets.csv")

# Group by language and business type
grouped = df2.groupby(['language', 'business_type']).size().reset_index(name='count')

# Find the language with the most problems for each business type
most_problems_by_type = grouped.loc[grouped.groupby('business_type')['count'].idxmax()]

# Find the language with the most problems related to IT services
it_services_language = most_problems_by_type[most_problems_by_type['business_type'] == 'IT Services']['language'].iloc[0]

# Find the language with the most problems related to Tech Online Store
tech_store_language = most_problems_by_type[most_problems_by_type['business_type'] == 'Tech Online Store']['language'].iloc[0]


# Find the language with the fewest problems overall
language_counts = df2['language'].value_counts()
least_problems_language = language_counts.index[-1]

print(f"Language with most problems related to IT Services: {it_services_language}")
print(f"Language with most problems related to Tech Online Store: {tech_store_language}")
print(f"Language with the fewest problems overall: {least_problems_language}")
```
